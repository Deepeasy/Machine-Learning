```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv("austin_weathernew.csv")
```

```python
data1.info()
```

```python
X=data1.drop(["PrecipitationSumInches"],axis=1)
```

```python
Y=data1["PrecipitationSumInches"]
```

```python
data1
```

```python
#reshaping Y
Y=Y.values.reshape(-1,1)
Y
```

```python
days_index=798 #random
days=[i for i in range (Y.size)]
```

```python
#Linear Regression
clf=LinearRegression()

clf.fit(X,Y)
```

```python
print("The precipitation trend graph")

plt.scatter(days, Y, color="g")
plt.scatter(days[days_index],Y[days_index], color="r")
plt.legend()
plt.title("Precipitation")
plt.xlabel("Days")
plt.ylabel("Preicipitation in Inches")
plt.show()
```

```python
x_vis=X.filter(["TempAvgF","DewPointAvgF","HumidityHighPercent","SeaLevelPressureAvgInches","VisibilityAvgMiles","WindAvgMPH"])
```

```python
data1.corr()
```

```python
print("The precipitation vs attributes trend graph")
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_vis[x_vis.columns.values[i][:100]], color='g')
    plt.scatter(days[days_index],x_vis[x_vis.columns.values[i]][days_index], color='r')
    plt.title(x_vis.columns.values[i])
plt.show()
```
